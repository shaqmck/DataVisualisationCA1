```python
#Imports
import os
import numpy as np
import pandas as pd
```

```python
#Get The directory and assaign file
os.getcwd()
os.chdir('C:\\Users\shaqu\Desktop\datavis\ca1\DataVisualisationCA1') 
data = 'sharktank.csv'
```

```python
#Put data into pandas DataFrame and display , encoding changed otherwise the file does not read correctly. 
sharktank = pd.read_csv(data,encoding='ISO-8859-1')
sharktank.head()
```

```python
#As Discussed with Ajay, replacing the files empty cells with 0's
sharktank.replace(np.nan, 0, inplace = True)
```

```python
#As of season 7 we lose pitcher information and some select company websites also contain empty cells
#We replace those 0's with 'N/A' to ensure the columns remain as containing only strings.
sharktank['Pitchers State'] = sharktank['Pitchers State'].astype(str) 
sharktank['Pitchers State'].replace(str(0),'N/A', inplace = True)

sharktank['Pitchers City'] = sharktank['Pitchers City'].astype(str) 
sharktank['Pitchers City'].replace(str(0),'N/A', inplace = True)

sharktank['Entrepreneur Names'] = sharktank['Entrepreneur Names'].astype(str) 
sharktank['Entrepreneur Names'].replace(str(0),'N/A', inplace = True)

sharktank['Company Website'] = sharktank['Company Website'].astype(str) 
sharktank['Company Website'].replace(str(0),'N/A', inplace = True)
```
